```python
def load_data(file_name, col_list = ['text', 'user.location', 'lang', 'created_at']):
    #Function to load in weekly data with relevant columns
    
    # Read the data as pandas dataframe
    data = pd.read_csv(file_name, usecols = col_list, lineterminator='\n')
    
    return data

# Interested in the text of the tweet (text) and when it was tweeted (created_at), where it's from (user.location).
# As focus is UK, we select based on (user.location) and (lang) to help geolocate the tweet.
# Note that user.location is user input space hence we need to perform the following steps to try and filter the data.
```

## Dominic Cummings

```python
#Dominic Cummings Data Files Dates
DC_W1_start_date = pd.to_datetime("2020-05-26")
DC_W1_end_date = pd.to_datetime("2020-06-02")
DC_W2_start_date = pd.to_datetime("2020-06-03")
DC_W2_end_date = pd.to_datetime("2020-06-09")
```

```python
## Dominic Cummings Part 1

# Load Data
this_data = load_data("/Users/brandon/Documents/UCL/Project/COVID Project/Data/dump1.csv")
print("dump1 contains", len(this_data), "tweets")

# Convert date column
this_data['created_at'] = pd.to_datetime(this_data.created_at).dt.normalize() ### CHECK

# Filter on date
this_data = this_data.loc[(this_data['created_at'].dt.date >= DC_W1_start_date) & (this_data['created_at'].dt.date <= DC_W1_end_date)]
print("dump1 contains", len(this_data), "tweets relevant for week 1")                                                                                   

# Preview
this_data.head(5)                                    
```

```python
this_data.to_pickle("DC_Data_Part1")
```
